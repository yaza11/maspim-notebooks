# Image algorithms

## Sample area detection
This is done in multiple steps
* binarization using Otsu thresholding to distinguish fore- from background pixels 
* use an optimizer to maximize the contrast with a box compared to the outside (first gif)
* use the box from before to find the contour of a multi-scaled median-filtered image (second gif + picture)







## Laminae detection
* First, the grayscale image is preprocessed to reduce noisy pixels
* Then, the foreground pixels are thresholded using an adaptive mean filter with a filter size corresponding to a multiple of the average expected thickness of a layer according to the age model (if one is present) or the combination of multi-scale median filters
* If no tilt correction is performed, a distorted rectangle is fitted to the thresholded image from the previous step. Otherwise only the width of the layers is optimized







## Descriptor and tilt correction
The descriptor is designed to describe images in terms of the parameters
* angle
* width
* phase
* and quality.
This is achieved by convolving the image with different wavelets. By default the wavelets are rectangular. The parameters of the wavelet with the highest coefficient are assigned to the pixel. Most importantly this allows assigning an angle to every pixel.






After some filtering, averaging and interpolating the streamlines are corrected from the angles which are used to define a deformation field that induces a transformation.
